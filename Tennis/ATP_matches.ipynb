```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
imports=list(range(2000,2018))
```

```python
for i,el in enumerate(imports):
    url = 'data/atp_matches_'+str(el)+'.csv'
    atp=pd.read_csv(url)
    if i == 0 : final = atp
    else : final=final.append(atp)
    print (len(final), len(atp))
```

```python
cols = final.columns
final=final.reset_index(drop=True)
```

```python
def lookup(s):
    """
    A fast approach to datetime parsing.Rather than re-parse these, we store all unique
    dates, parse them, and use a lookup to convert all dates.
    """
    dates = {date:pd.to_datetime(date, format='%Y%m%d', errors='coerce') for date in s.unique()}
    return s.map(dates)
```

```python
final.tourney_date=lookup(final.tourney_date)
```

```python
final=final[-final.tourney_date.isna()]
```
